```python

# Step 6: Adding New Columns
# Create a new column 'Happiness Category'
def categorize_happiness(score):
    if score < 4:
        return "Low"
    elif score <= 6:
        return "Medium"
    else:
        return "High"
```

```python
df['Happiness Category'] = df['score'].apply(categorize_happiness)
sorted_df.head()
```

```python
# Step 7: Data Visualizations
# 1. Bar Plot: Top 10 happiest countries
top_10_happiest = df.nlargest(10, 'score')
plt.figure(figsize=(10, 6))
sns.barplot(x='score', y='Country name', data=top_10_happiest, palette='viridis')
plt.title("Top 10 Happiest Countries by score")
plt.xlabel("Happiness score")
plt.ylabel("Country")
plt.show()
```

```python
# 2. Line Plot: Top 10 unhappiest countries
top_10_unhappiest = df.nsmallest(10, 'score')
plt.figure(figsize=(10, 6))
sns.lineplot(x='score', y='Country name', data=top_10_unhappiest, marker='o')
plt.title("Top 10 Unhappiest Countries by score")
plt.xlabel("Happiness score")
plt.ylabel("Country name")
plt.show()
```

```python
# 3. Histogram for the "Score" column
plt.figure(figsize=(8, 6))
sns.histplot(df['score'], bins=20, kde=True, color='blue')
plt.title("Distribution of Happiness scores")
plt.xlabel("Happiness score")
plt.ylabel("Frequency")
plt.show()
```

```python
# Task 3: Outlier Detection
# 1. Identify outliers based on Score and GDP per Capita using the 1.5x IQR rule
q1_gdp, q3_gdp = south_asia_df['Log GDP per capita'].quantile([0.25, 0.75])
iqr_gdp = q3_gdp - q1_gdp
gdp_lower = q1_gdp - 1.5 * iqr_gdp
gdp_upper = q3_gdp + 1.5 * iqr_gdp
```

```python
q1_score, q3_score = south_asia_df['score'].quantile([0.25, 0.75])
iqr_score = q3_score - q1_score
score_lower = q1_score - 1.5 * iqr_score
score_upper = q3_score + 1.5 * iqr_score
```

```python
outliers = south_asia_df[
    (south_asia_df['Log GDP per capita'] < gdp_lower) |
    (south_asia_df['Log GDP per capita'] > gdp_upper) |
    (south_asia_df['score'] < score_lower) |
    (south_asia_df['score'] > score_upper)
]
```

```python
# 2. Scatter plot highlighting outliers
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_df, label='Normal Data', color='blue')
sns.scatterplot(x='Log GDP per capita', y='score', data=outliers, label='Outliers', color='red', marker='X', s=100)
plt.title("Log GDP per capita vs. score (Outliers Highlighted)")
plt.xlabel("Log GDP per capita")
plt.ylabel("Happiness Score")
plt.legend()
plt.show()
```

```python
# Qs no 6
# Detect outliers in both regions
def detect_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return df[(df[column] < lower) | (df[column] > upper)]

outliers_sa = detect_outliers(south_asia_df, 'score')
outliers_me = detect_outliers(middle_east_df, 'score')
```

```python
# Scatter plot highlighting outliers
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_df, label='South Asia')
sns.scatterplot(x='Log GDP per capita', y='score', data=middle_east_df, label='Middle East')
sns.scatterplot(x='Log GDP per capita', y='score', data=outliers_sa, label='SA Outliers', color='red', marker='X', s=100)
sns.scatterplot(x='Log GDP per capita', y='score', data=outliers_me, label='ME Outliers', color='purple', marker='X', s=100)
plt.title("Log GDP per Capita vs. Score (Outliers Highlighted)")
plt.xlabel("Log GDP per Capita")
plt.ylabel("Happiness Score")
plt.legend()
plt.show()
```
